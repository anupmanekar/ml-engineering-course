# Graph Plotting with Arrays

## Matplotlib - Plotting library

#### Matplotlib is generally used for generating 2D and 3D graphs from set of numbers generated by numpy. Lets try with generating a 2D graph

```python
import numpy as np
import matplotlib .pylab as plt
a = np.random.rand(10)
plt.plot(a)
plt.show()
```

#### Now lets generate a 3D graph

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pylab as plt
import numpy as np

x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z)
plt.show()
```
